```python
# === Metrics from Conditional Diffusion Model ===

pokie_scores = [0.6638, 0.6563, 0.6558, 0.6590, 0.6614, 0.6632, 0.6604, 0.6595, 0.6621, 0.6577]
mmd_scores = [0.010510, 0.007826, 0.008493, 0.009691, 0.006821, 0.016327, 0.008206, 0.010795, 0.012075, 0.010393]
wasserstein_scores = [2.553170, 1.804791, 2.672473, 2.578355, 2.451918,
                      2.625246, 2.485911, 2.332111, 2.602027, 2.392323]

# Estimate the missing MMD score (for digit 4) as the mean of existing ones (optional)
import numpy as np
mmd_missing = np.mean(mmd_scores)  # Simple imputation for demo
mmd_scores.insert(4, mmd_missing)

# === Compute Means ===
avg_pokie = sum(pokie_scores) / len(pokie_scores)
avg_mmd = sum(mmd_scores) / len(mmd_scores)
avg_wass = sum(wasserstein_scores) / len(wasserstein_scores)

# === Print Results ===
print(f"Average Pokie Score      : {avg_pokie:.4f}")
print(f"Average MMD Score        : {avg_mmd:.6f}")
print(f"Average Wasserstein Dist : {avg_wass:.6f}")
```
